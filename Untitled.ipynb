```python
import pandas as pd
```

```python
df = pd.read_pickle('data/vinyl_collection.pkl')
```

```python
color_mapping = {
    'Blue': ['blue', 'aqua', 'ocean'],
    'Purple': ['purple', 'violet'],
    'Green': ['green', 'emerald', 'glow in the dark'],
    'Yellow': ['yellow', 'gold'],
    'Pink': ['pink'],
    'Red': ['red', 'maroon'],
    'Orange': ['orange', 'tangerine'],
    'White': ['white', 'cream'],
    'Brown': ['brown', 'chestnut'],
    'Clear': ['clear'],
    'Picture Disc': ['picture disc'],
    'Gray': ['gray', 'grey'],
    'Black': ['black']}
```

```python
def check_picture_disc(row):
    """Replaces pressing color with 'Picture Disc' when applicable."""
    if 'Picture Disc' in row['description']:
        return ['Picture Disc']
    else:
        return row['pressing']
```

```python
# New dataframe of pressing colors and wants/haves
df_colors = df[['id', 'pressing', 'description']].copy()

# Ratio of wants to haves: the higher the number, more people want it


df_colors['pressing'] = df_colors.apply(check_picture_disc, axis=1)
```

```python
df_colors
```

```python
def map_colors(row):
    mapped_colors = []
    for color in row['color']:
        for key, values in color_mapping.items():
            if any(val.lower() in color.lower() for val in values):
                mapped_colors.append(key)
    return mapped_colors
```

```python
df_colors['new_color'] = df_colors.apply(map_colors, axis=1)
```
